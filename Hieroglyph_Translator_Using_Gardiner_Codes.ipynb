## Local Inference on GPU
Model page: https://huggingface.co/AnushS/Hieroglyph-Translator-Using-Gardiner-Codes

⚠️ If the generated code snippets do not work, please open an issue on either the [model repo](https://huggingface.co/AnushS/Hieroglyph-Translator-Using-Gardiner-Codes)
			and/or on [huggingface.js](https://github.com/huggingface/huggingface.js/blob/main/packages/tasks/src/model-libraries-snippets.ts) 🙏

```python
# Load model directly
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM

tokenizer = AutoTokenizer.from_pretrained("AnushS/Hieroglyph-Translator-Using-Gardiner-Codes")
model = AutoModelForSeq2SeqLM.from_pretrained("AnushS/Hieroglyph-Translator-Using-Gardiner-Codes")
```

```python
gardiner_codes_examples = [
    "A4 A5 G4 H9 P3",
    "N35 V30 n H8",
    "D4 D5 Z1 G1"
]

for codes in gardiner_codes_examples:
    text = f"Translate hieroglyph gardiner code sequence to English: {codes}"
    inputs = tokenizer(text, return_tensors="pt").input_ids
    outputs = model.generate(inputs, max_new_tokens=40, do_sample=True, top_k=30, top_p=0.95)
    translated_keywords = str(tokenizer.decode(outputs[0], skip_special_tokens=True))
    print(f"Gardiner Codes: {codes}")
    print(f"Translation: {translated_keywords}\n")
```

```python
text = "Translate hieroglyph gardiner code sequence to English: A4 A5 A1 B6 F8"

inputs = tokenizer(text, return_tensors="pt").input_ids
outputs = model.generate(inputs, max_new_tokens=40, do_sample=True, top_k=30, top_p=0.95)
translated_keywords = str(tokenizer.decode(outputs[0], skip_special_tokens=True))

print(translated_keywords)
```
